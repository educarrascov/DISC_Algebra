# Funciones de Optimización (Benchmark):

1. **Ackley Function:** Presenta un paisaje de optimización complejo con un mínimo global rodeado de numerosos mínimos locales. Es conocida por su superficie de valles y picos.

2. **Rastrigin Function:** Caracterizada por un paisaje ondulado con muchos mínimos locales, lo que la hace desafiante para la optimización. Tiene un mínimo global en el centro.

3. **Beale Function:** Una función bidimensional con un valle estrecho y un único mínimo global. Es utilizada para probar la precisión de los algoritmos en superficies estrechas y curvas.

4. **Goldstein-Price Function:** Posee un paisaje de búsqueda complejo con múltiples mínimos locales y un mínimo global distintivo. Es una prueba desafiante para la localización de mínimos globales.

5. **Himmelblau Function:** Tiene un paisaje con varios mínimos locales y cuatro mínimos globales idénticos. Es útil para probar la capacidad de los algoritmos para encontrar todos los mínimos globales posibles.

6. **Lévi Function:** Ofrece un paisaje de optimización irregular con un único mínimo global. Es conocida por sus características onduladas y es desafiante para algoritmos que buscan evitar mínimos locales.

7. **Schwefel Function:** Conocida por su mínimo global en el límite del dominio de búsqueda, presenta un desafío único para los algoritmos en encontrar mínimos en los bordes del espacio de búsqueda.

8. **Booth Function:** Una función de optimización simple con un único mínimo global. Es a menudo usada como una prueba inicial para algoritmos debido a su simplicidad.

9. **Rosenbrock Function:** Conocida como la función del "valle de Rosenbrock", presenta un largo y estrecho valle parabólico con un mínimo global. Es famosa por ser un desafío para algoritmos que deben seguir el valle hasta encontrar el mínimo.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definición de las funciones de prueba de optimización
def ackley_function(x, y):
    return -20.0 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2))) - np.exp(0.5 * (np.cos(2.0 * np.pi * x) + np.cos(2.0 * np.pi * y))) + np.e + 20

def rastrigin_function(x, y):
    return 10 * 2 + (x**2 - 10 * np.cos(2 * np.pi * x)) + (y**2 - 10 * np.cos(2 * np.pi * y))

def beale_function(x, y):
    return (1.5 - x + x * y)**2 + (2.25 - x + x * y**2)**2 + (2.625 - x + x * y**3)**2

def goldstein_price_function(x, y):
    return (1 + (x + y + 1)**2 * (19 - 14*x + 3*x**2 - 14*y + 6*x*y + 3*y**2)) * \
           (30 + (2*x - 3*y)**2 * (18 - 32*x + 12*x**2 + 48*y - 36*x*y + 27*y**2))

def himmelblau_function(x, y):
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2

def levi_function(x, y):
    return np.sin(3 * np.pi * x)**2 + ((x - 1)**2) * (1 + np.sin(3 * np.pi * y)**2) + ((y - 1)**2) * (1 + np.sin(2 * np.pi * y)**2)

def schwefel_function(x, y):
    return 418.9829 * 2 - (x * np.sin(np.sqrt(np.abs(x))) + y * np.sin(np.sqrt(np.abs(y))))

def booth_function(x, y):
    return (x + 2*y - 7)**2 + (2*x + y - 5)**2

def rosenbrock_function(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Creación de una grilla de puntos para evaluar las funciones
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
x_grid, y_grid = np.meshgrid(x, y)

# Lista de funciones para iterar y graficar
functions = [ackley_function, rastrigin_function, beale_function, goldstein_price_function,
             himmelblau_function, levi_function, schwefel_function, booth_function, rosenbrock_function]

# Nombres de las funciones para títulos
function_names = ['Función Ackley', 'Función Rastrigin', 'Función Beale', 'Función Goldstein-Price', 
                  'Función Himmelblau', 'Función Lévi', 'Función Schwefel', 'Función Booth', 'Función Rosenbrock']

# Configuración de la figura para 9 subplots
fig = plt.figure(figsize=(20, 20))

# Crear subgráficos para cada función
for i, function in enumerate(functions, start=1):
    ax = fig.add_subplot(3, 3, i, projection='3d')
    z_grid = function(x_grid, y_grid)
    ax.plot_surface(x_grid, y_grid, z_grid, cmap='viridis', alpha=0.8)
    ax.set_title(function_names[i - 1], fontsize=20)

# Ajustar el espaciado entre los subgráficos
plt.tight_layout()
plt.show()
```
